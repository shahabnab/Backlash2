```python
import numpy as np
import matplotlib.pyplot as plt
import math
import pprint

import cma
```

```python
data = np.array([line.split() for line in data.strip().split('\n')], dtype='f8')
t, v = data[:, 0], data[:, 1]

# plt.rcParams['figure.figsize']
plt.rcParams['figure.figsize'] = [12, 5]
# %matplotlib inline

plt.plot(t, v, 'b-o')
plt.xlabel('time (s)')
plt.ylabel('Speed (rpm)')
plt.legend(['data'])
plt.show()
```

```python
#parameters value (empirically found)
A = 0.5
t0  = 0.19
tau = 9
omega = 2*np.pi*26
c = 99.98 # np.mean(v)
t1 = 0.39
t2 = 0.2


def media(v):
   return sum(v)/len(v)

def RMSE(v,v_hat):
      return np.sqrt(((v-v_hat)**2).mean())

def step(x):
    return 1*(x>0)

# A rectangular step function on [0, 1]:
def rect(x):
    return 0.5*(np.sign(x) + np.sign(1 - x))


# Interpolating func: 
# v_hat = A*step(t-t2) *step(t1-t) *np.exp(-1*(t-t2)*tau) *np.sin(omega*(t-t0)) +c
# par = [A, t0, tau, omega, c, t1, t2]

# 7 parameters
def v_hat(par):
    return par[0] *step(t-par[6]) *step(par[5]-t) *np.exp(-1*(t-par[6])*par[2]) *np.sin(par[3]*(t-par[1]))+par[4]

def obj(par):
    return RMSE(v,v_hat(par))



# with just 2 unknown parameters [tau, t1,]
def obj2(b):
    return RMSE(v,v_hat([A, t0, b[0], omega, c, b[1], t2]))

# with just 3 unknown parameters [A, tau, t1,]
def obj3(b):
    return RMSE(v,v_hat([b[0], t0, b[1], omega, c, b[2], t2]))
```
